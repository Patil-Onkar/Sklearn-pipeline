```python
def change_colour(val):
    #print(val)
    if val['per']>=95:
        return ['background-color: #8eed11']*len(val)
    elif val['per']>=70 and val['per']<=95:
        return ['background-color: #f7b052']*len(val)
    elif val['per']<70: 
        return ['background-color: #8eed11']*len(val)
```

```python
df=df[:10]
```

```python
k=df.style.set_properties(**{'border': '1px black solid !important',
                                                          'text-align': 'Left',
                                                        'font-size': '12pt',
                                                        'font-weight': 'bold',
                                                        'width': '800px'}).hide_index().set_table_styles([{'selector': 'thead', 'props': [('background-color', 'black'), ('color', 'cyan')] }]).\
                                        apply(change_colour,axis=1)#.apply(change_fontcolour_info)
                        
#change_colour

#df_style.render()
```
